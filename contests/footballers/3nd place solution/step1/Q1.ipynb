```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('train.csv')
train
```

```python
train.playerId.nunique()
```

```python
train[train.outcome == 'گُل'].playerId.value_counts()
```

```python
rate = []
for player in train.playerId.unique():
    total = len(train[train.playerId == player])
    goal = len(train[train.playerId == player][train.outcome == 'گُل'])
    rate.append(goal/total)
```

```python
temp = pd.DataFrame({'player': train.playerId.unique(), 'rate': rate})
temp.sort_values(['rate'], ascending=False)
```
